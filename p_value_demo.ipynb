```python
data = np.array([980, 995, 1005, 1020, 990, 1010, 1002, 1008, 997, 1003])

hypt_default = 1000

alpha = 0.05
t_stat,p_value = stats.ttest_1samp(data,hypt_default)

print("P-value:", p_value)
```

```python
import numpy as np
from scipy.stats import describe, norm
# Generate some random data from a normal distribution
data = norm.rvs(loc=0, scale=1, size=100)
print(data)

print()

# Calculate descriptive statistics using describe
desc = describe(data)
print(desc)

print()

# Print the descriptive statistics
print("Descriptive Statistics:")
print("Mean:", desc.mean)
print("Minimum:", desc.minmax[0])
print("Maximum:", desc.minmax[1])
print("Variance:", desc.variance)
print("Skewness:", desc.skewness)
print("Kurtosis:", desc.kurtosis)
```

```python
#new optimized route is actually reducing delivery times

from scipy.stats import ttest_ind

old_route = [31, 29, 35, 30, 33, 32]
new_route = [28, 27, 29, 26, 30, 28]

alpha = 0.01

my_stat,p_value = ttest_ind(old_route,new_route)

print("Stats ", my_stat, "P_value is ",p_value)

if p_value < alpha:
  print("New route is faster")
else:
  print("Old route is faster")
```
